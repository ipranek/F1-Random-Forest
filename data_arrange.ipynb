```python
import os
import pandas as pd

DIR_PATH = "/Users/ipekoner/Desktop/f1-ml/f1-data"

def load_and_preview_csv(file_name, dir_path):
    file_path =os.path.join(dir_path, file_name)
    df = pd.read_csv(file_path)
    return df.head(), df.info(), df.describe()

filenames = [f for f in os.listdir(DIR_PATH) if f.endswith(".csv")]

data_overview = {}

for file in filenames:
    data_overview[file] = load_and_preview_csv(file, DIR_PATH)

print(data_overview.keys())

print(os.listdir(DIR_PATH))
```

```python
import pandas as pd

results = pd.read_csv(os.path.join(DIR_PATH, "results.csv"))
drivers = pd.read_csv(os.path.join(DIR_PATH, "drivers.csv"))

merged = results.merge(drivers, on="driverId", how="inner")

final_df = merged[["driverId", "forename", "surname","position", "grid"]]

final_df
```

```python
races = pd.read_csv(os.path.join(DIR_PATH, "races.csv"))

merged = results.merge(drivers, on="driverId", how="inner") #merging drivers to results with the key as driverId 

merged = merged.merge(races[["raceId", "name", "date"]], on="raceId", how="inner") #taking the merged and merging races to it with the selected columns with the key as raceId

final_df = merged[["raceId", "name", "driverId", "forename", "surname", "position", "grid", "constructorId"]]

final_df
```

```python
merged["grid"] = pd.to_numeric(merged["grid"], errors = "coerce")
merged["position"] = pd.to_numeric(merged["position"], errors = "coerce")

merged["delta"] = merged["grid"] - merged["position"]

print(merged[["raceId", "name", "driverId", "forename", "surname", "grid", "position",  "delta"]])

final_df["delta"] = merged["grid"] - merged["position"]

final_df
```

```python
teams = pd.read_csv(os.path.join(DIR_PATH, "constructors.csv"))
team_points = pd.read_csv(os.path.join(DIR_PATH, "constructor_results.csv"))

team_points = team_points.rename(columns={"points": "constructor_points"})

merged = merged.merge(team_points, on=["raceId", "constructorId"], how = "left") #left so that every drivers team keeps their points and the rows dont drop
merged = merged.merge(teams[["constructorId", "name"]], on ="constructorId", how ="left", suffixes = ("", "_constructor"))

print(merged[["raceId", "name", "driverId", "forename", "surname", "grid", "position",  "delta", "name_constructor", "constructor_points"]])

```
